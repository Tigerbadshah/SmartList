```python
import spacy
import math

class SmartList():
    def __init__(self,campareTo):
        if len(campareTo) > 0:
            self.campareTo = campareTo
        else:
            self.campareTo = None
            print("Empty List was Given")
            
    def campare(self,campareWith, getResult=True):
        if self.campareTo == None:
            print("Empty List was Given")
        elif len(str(campareWith)) <= 0:
            print("Camparing Text Couldn't Pass through")
        else:
            #campareWith.similarity(nlp(value))*100
            nlp = spacy.load("en_core_web_md")
            campareWith = nlp(campareWith)
            similarity = []
            for value in self.campareTo:
                similarity.append(campareWith.similarity(nlp(value))*100)
            if getResult == True:
                i = 0
                topListPercent = 0
                topListIndex = 0
                while i < len(similarity):
                    if topListPercent < similarity[i]:
                        topListPercent = similarity[i]
                        topListIndex = i
                    i=i+1
                return(similarity,[self.campareTo[topListIndex],topListIndex,topListPercent])
            else:
                return(similarity)
        
    def getList(self):
        return(self.campareTo)
    
    def addToList(self,toAdd,place=False):
        if place == False:
            self.campareTo.append(toAdd)
        else:
            self.campareTo.insert(place, toAdd)
    
    def removeFromList(self,toRemove):
        temp = False
        for value in self.campareTo:
            if toRemove == value:
                temp = True
        if temp == True:
            self.campareTo.remove(toRemove)
        else:
            print('Element not in List')
```
